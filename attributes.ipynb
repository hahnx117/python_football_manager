```python
# Import required packages
import numpy as np
import pandas as pd
#import plotly.express as px
#from easygui import *
import matplotlib.pyplot as plt
```

```python
DATA_DIR=f"/home/david/Documents/python_football_manager"
STATS_FILE=f"stats.html"
ATTRIBUTES_FILE=f"attributes.html"
```

```python
# Set DataFrame limits
pd.set_option('display.max_columns',50)
pd.set_option('display.max_rows',10)
```

```python
attributes_df = pd.read_html(f"{DATA_DIR}/{ATTRIBUTES_FILE}")
stats_df = pd.read_html(f"{DATA_DIR}/{STATS_FILE}")

attributes_df = attributes_df[0]
stats_df = stats_df[0]

attributes_df["Name"] = attributes_df["Name"].astype(str)
stats_df["Name"] = stats_df["Name"].astype(str)
```

```python
attributes_dict = {
    'Acceleration':'Acc',
    'Work Rate':'Wor',
    'Vision':'Vis',
    'Goalkeeper Throwing':'Thr',
    'Technique':'Tec',
    'Teamwork':'Tea',
    'Tackling':'Tck',
    'Strength':'Str',
    'Stamina':'Sta',
    'TRO':'TRO',
    'Reflexes':'Ref',
    'Tendency to Punch':'Pun',
    'Positioning':'Pos',
    'Penalties':'Pen',
    'Passing':'Pas',
    'Pace':'Pac',
    'One on Ones':'1v1',
    'Off the Ball':'OtB',
    'Natural Fitness':'Nat',
    'Marking':'Mar',
    'Long Throws':'L Th',
    'Long Shots':'Lon',
    'Leadership':'Ldr',
    'Kicking':'Kic',
    'Jumping':'Jum',
    'Heading':'Hea',
    'Handling':'Han',
    'Free Kicks':'Fre',
    'Flair':'Fla',
    'First Touch':'Fir',
    'Finishing':'Fin',
    'Eccentricity':'Ecc',
    'Dribbling':'Dri',
    'Determination':'Det',
    'Decisions':'Dec',
    'Crossing':'Cro',
    'Corners':'Cor',
    'Concentration':'Cnt',
    'Composure':'Cmp',
    'Communication':'Com',
    'Command of Area':'Cmd',
    'Bravery':'Bra',
    'Balance':'Bal',
    'Anticipation':'Ant',
    'Agility':'Agi',
    'Aggression':'Agg',
    'Aerial Ability':'Aer',
    'Name': 'Name',
}
```

```python
stats_df.columns.values
```

```python
# Let's try to create attribute lists for the players of general positions
goalkeeper_attributes = ["Name", "Handling", "Reflexes", "One on Ones", "Aerial Ability", "Communication", "Command of Area", "Concentration", "Decisions", "Teamwork", "Work Rate", "Kicking"]
central_defender_attributes = ["Name", "Tackling", "Marking", "Positioning", "Jumping", "Strength", "Bravery", "Concentration", "Decisions", "Teamwork", "Work Rate"]
full_back_attributes = ["Name", "Marking", "Pace", "Acceleration", "Passing", "Crossing"]
defensive_mid_attributes = ["Name", "Tackling", "Marking", "Positioning", "Passing", "Teamwork"]
central_mid_attributes = ["Name", "Passing", "Teamwork", "Work Rate", "Decisions", "Concentration", "Bravery"]
attacking_mid_attributes = ["Name", "Passing", "Dribbling", "Finishing", "Pace", "Acceleration", "Agility", "Balance", "Flair", "Technique", "Decisions", "Concentration", "Bravery"]
winger_attributes = ["Name", "Pace", "Acceleration", "Agility", "Balance", "Dribbling", "Crossing", "Finishing", "Flair", "Technique", "Decisions", "Concentration", "Bravery"]
striker_attributes = ["Name", "Finishing", "Heading", "Pace", "Acceleration", "Agility", "Balance", "Dribbling", "Flair", "Technique", "Decisions", "Concentration", "Bravery"]

goalkeeper_stats = ["Name", "Saves/90", "Cln/90"]
central_defender_stats = ["Name", "Cln/90", "Tck/90"]

# now create a dataframe using a dict to translate for attribute names
# goalkeeper
goalkeeper_df = attributes_df[[ attributes_dict[x] for x in goalkeeper_attributes ]].copy()
goalkeeper_stats_df = stats_df[[ x for x in goalkeeper_stats]].copy()
goalkeeper_df['Att_Total'] = goalkeeper_df.loc[:,].sum(axis=1, numeric_only = True)

print(goalkeeper_stats_df)

goalkeeper_df.join(goalkeeper_stats_df, on="Name", how="right")

# central_defender
central_defender_df = attributes_df[[ attributes_dict[x] for x in central_defender_attributes ]].copy()
central_defender_df['Att_Total'] = central_defender_df.loc[:,].sum(axis=1, numeric_only = True)

# full_back
full_back_df = attributes_df[[ attributes_dict[x] for x in full_back_attributes ]].copy()
full_back_df['Att_Total'] = full_back_df.loc[:,].sum(axis=1, numeric_only = True)

# defensive_mid
defensive_mid_df = attributes_df[[ attributes_dict[x] for x in defensive_mid_attributes ]].copy()
defensive_mid_df['Att_Total'] = defensive_mid_df.loc[:,].sum(axis=1, numeric_only = True)

# central_mid
central_mid_df = attributes_df[[ attributes_dict[x] for x in central_mid_attributes ]].copy()
central_mid_df['Att_Total'] = central_mid_df.loc[:,].sum(axis=1, numeric_only = True)

# attacking_mid
attacking_mid_df = attributes_df[[ attributes_dict[x] for x in attacking_mid_attributes ]].copy()
attacking_mid_df['Att_Total'] = attacking_mid_df.loc[:,].sum(axis=1, numeric_only = True)

# winger
winger_df = attributes_df[[ attributes_dict[x] for x in winger_attributes ]].copy()
winger_df['Att_Total'] = winger_df.loc[:,].sum(axis=1, numeric_only = True)

# striker
striker_df = attributes_df[[ attributes_dict[x] for x in striker_attributes ]].copy()
striker_df['Att_Total'] = striker_df.loc[:,].sum(axis=1, numeric_only = True)
```
